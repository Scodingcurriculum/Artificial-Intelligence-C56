```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Sample Data: Animal sounds (bark=0, meow=1, moo=2) with features (frequency and duration)
data = {
    'Frequency': [300, 450, 500, 250, 600, 350, 400, 100],
    'Duration': [1, 0.5, 0.8, 1.2, 0.7, 0.9, 1, 1.5],
    'Sound': [0, 1, 1, 0, 2, 1, 2, 0]  # Labels (bark=0, meow=1, moo=2)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Split data into features (X) and labels (y)
X = df[['Frequency', 'Duration']]
y = df['Sound']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier()

# Train the model
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)

# Test the model with new animal sound data
new_sound = [[320, 1.1]]  # Example of a new sound (Frequency=320, Duration=1.1)
predicted_sound = clf.predict(new_sound)

# Mapping prediction to animal
animal_sounds = {0: 'Bark', 1: 'Meow', 2: 'Moo'}
print("Predicted Sound:", animal_sounds[predicted_sound[0]])

# Plot the original data points
plt.figure(figsize=(10, 6))
plt.scatter(df['Frequency'], df['Duration'], c=df['Sound'], cmap='viridis', s=100, edgecolor='k')

# Highlight the new sound prediction
plt.scatter(new_sound[0][0], new_sound[0][1], c='red', s=200, label='New Sound (Predicted)', edgecolor='k')

# Add labels and title
plt.title('Animal Sounds Based on Frequency and Duration')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Duration (seconds)')
plt.colorbar(label='Sound (0: Bark, 1: Meow, 2: Moo)')
plt.legend()

# Show the plot
plt.show()
```

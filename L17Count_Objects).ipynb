```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload the image file
uploaded = files.upload()  # Use this to upload the image file

# Read and process the image
image = cv2.imread('1.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur
blur = cv2.GaussianBlur(gray, (11, 11), 0)

# Canny edge detection
canny = cv2.Canny(blur, 30, 150, 3)

# Dilate the image to close gaps
dilated = cv2.dilate(canny, (1, 1), iterations=0)

# Find contours
(cnt, hierarchy) = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# Convert BGR to RGB
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Draw contours on the image
cv2.drawContours(rgb, cnt, -1, (0, 255, 0), 2)

# Count the number of coins detected
coin_count = len(cnt)

# Display the count on the image
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(rgb, f'Coins: {coin_count}', (10, 30), font, 1, (255, 0, 0), 2, cv2.LINE_AA)

# Show the number of coins detected
print("Coins in the image: ", coin_count)

# Display the image with contours and coin count
plt.imshow(rgb)
plt.axis('off')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Sample data: student ratings for various activities (scale: 1 to 5)
data = {
    'Reading': [5, 4, 2, 1, 3, 5, 4, 2, 3, 1],
    'Drawing': [2, 3, 4, 5, 1, 3, 2, 4, 5, 4],
    'Sports': [3, 2, 5, 4, 4, 1, 5, 5, 3, 2],
    'Singing': [1, 2, 3, 5, 4, 2, 1, 3, 5, 4]
}

# Convert data into a DataFrame
df = pd.DataFrame(data)

# Normalize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Apply K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(df_scaled)

# Add cluster labels to the DataFrame
df['Cluster'] = clusters

# Display the clustered data
print("Clustered Data:")
print(df)

# Visualize the clusters using a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Reading'], df['Sports'], c=clusters, cmap='viridis')
plt.title('Student Clusters Based on Activity Preferences')
plt.xlabel('Reading Rating')
plt.ylabel('Sports Rating')
plt.colorbar(label='Cluster')
plt.show()
```

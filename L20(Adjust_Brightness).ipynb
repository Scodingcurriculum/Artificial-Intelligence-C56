```python
# Step 1: Install OpenCV
!pip install opencv-python

# Step 2: Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 3: Function to adjust brightness
def adjust_brightness(image, beta):
    """
    Adjusts the brightness of an image.
    Parameters:
        image: The input image.
        beta: The brightness adjustment value. Positive values increase brightness, negative values decrease it.
    Returns:
        The brightness-adjusted image.
    """
    # Clip the values to stay within the valid range (0-255)
    adjusted = cv2.convertScaleAbs(image, alpha=1, beta=beta)
    return adjusted

# Step 4: Load an image (You can upload an image or use a URL)
from google.colab import files
uploaded = files.upload()  # Run this line to upload an image from your local machine

# Assuming you uploaded an image named 'your_image.jpg'
image_path = "your_image.jpg"  # Replace with your image file name
original_image = cv2.imread("image.jpg")

# Step 5: Adjust brightness
beta_value = 50  # Change this value to increase or decrease brightness
bright_image = adjust_brightness(original_image, beta_value)

# Step 6: Display the original and adjusted images
# Convert BGR to RGB for displaying with matplotlib
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
bright_image_rgb = cv2.cvtColor(bright_image, cv2.COLOR_BGR2RGB)

# Plot the images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(original_image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(bright_image_rgb)
plt.title('Brightness Adjusted Image')
plt.axis('off')

plt.show()
```
